```python
from pyspark import SparkContext
from operator import add
```

```python
sc = SparkContext("local[*]","customer_data")
```

```python
rdd1 = sc.textFile("customerorders-201008-180523.csv")
```

```python
rdd1.collect()
```

```python
rdd2 = rdd1.map(lambda x: (x.split(',')[0],float(x.split(',')[2])))
```

```python
rdd2.collect()
```

```python
rdd3 = rdd2.reduceByKey(add)
```

```python
rdd3.collect()
```

```python
rdd4 = rdd3.sortByKey()
```

```python
rdd4.collect()
```

```python
#using a function
def parseLine(line):
    vals = line.split(',')
    vals_a = float(vals[0])
    vals_b = int(vals[0])
    return (vals_a,vals_b)
```

```python
rddx = rdd1.map(parseLine)
```

```python
rddx.collect()
```
